## Concept 2: Volume

### 1. The Math: "Counting the Votes"

Volume is the simplest data point in trading, but often the most ignored. It is the **total number of shares (or contracts)** traded during a specific time period (like one day).

- **Price Up + High Volume:** Strong conviction. Many people agree the price should be higher.
    
- **Price Up + Low Volume:** Weak conviction. The price moved, but "nobody was invited to the party." Itâ€™s more likely to reverse.
    

---

### 2. The Visualization: The "Support" Indicator

In charts, Volume is almost always shown as a **Bar Chart** at the bottom.

- A **Green Bar** usually means the price closed higher than it opened (Buying pressure).
    
- A **Red Bar** means the price closed lower (Selling pressure).
    

---

### 3. Let's look at the "Numbers"

VectorBT pulls volume automatically. You can compare it to the price to see if a move has "power."

```python
import vectorbt as vbt
import plotly.graph_objects as go

# 1. Download Data (includes Volume)
data = vbt.YFData.download('AAPL', period='3mo')
price = data.get('Close')
volume = data.get('Volume')

# 2. Calculate a Volume Moving Average (VMA)
# This helps us see if today's volume is "High" compared to the last 20 days
vma20 = vbt.MA.run(volume, window=20).ma

# 3. Visualize Price vs Volume
fig = vbt.make_subplots(rows=2, cols=1, shared_xaxes=True, 
                        vertical_spacing=0.05,
                        subplot_titles=("Price", "Volume with 20-day Average"))

fig.add_trace(go.Scatter(x=price.index, y=price, name='Price'), row=1, col=1)

# Add Volume Bars
fig.add_trace(go.Bar(x=volume.index, y=volume, name='Volume', marker_color='gray'), row=2, col=1)

# Add the Volume Moving Average to see "Spikes"
fig.add_trace(go.Scatter(x=vma20.index, y=vma20, name='Avg Volume', line=dict(color='orange')), row=2, col=1)

fig.show()
```

### 4. How to use Volume in a Strategy?

The most common way to use this in math is to look for **Volume Spikes**. If `Volume > (1.5 * Avg_Volume)`, it means something big is happening (like an earnings report or major news).

**The Logic:** * "I want to buy if the RSI is < 30 (**Oversold**) AND Volume is > Average (**High Interest**)."

### The Key takeaway for Volume:

Volume validates the price. If the price is crashing on low volume, it might just be a temporary dip. If itâ€™s crashing on huge volume, the "big players" are selling, and you should probably stay away!
