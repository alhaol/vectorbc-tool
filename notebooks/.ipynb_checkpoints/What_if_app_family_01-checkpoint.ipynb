```python
import vectorbt as vbt
price = vbt.YFData.download('BTC-USD',period='10y').get('Close')
pf = vbt.Portfolio.from_holding(price, init_cash=10000)
print(pf.total_profit())
```

```python
import vectorbt as vbt
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Define the UI Elements
ticker_input = widgets.Text(value='NVDA', description='Ticker:')
cash_slider = widgets.IntSlider(value=25000, min=1000, max=100000, step=1000, description='Init Cash:')
button = widgets.Button(description="Run Backtest", button_style='success')
output = widgets.Output()

# 2. Define the Logic
def run_backtest(b=None):
    with output:
        clear_output()
        try:
            print(f"Fetching data for {ticker_input.value}...")
            price = vbt.YFData.download(ticker_input.value, period='10y').get('Close')
            pf = vbt.Portfolio.from_holding(price, init_cash=cash_slider.value)
            
            print(f"--- Results for {ticker_input.value} ---")
            print(f"Total Profit: ${pf.total_profit():,.2f}")
            
            # VectorBT also has built-in interactive plots
            pf.plot().show() 
            
        except Exception as e:
            print(f"Error: {e}")

# 3. Connect and Display
button.on_click(run_backtest)
ui = widgets.VBox([widgets.HBox([ticker_input, cash_slider]), button])
display(ui, output)
```

```python
def simple_backtest(ticker, cash):
    price = vbt.YFData.download(ticker, period='3d').get('Close')
    pf = vbt.Portfolio.from_holding(price, init_cash=cash)
    return f"Total Profit: ${pf.total_profit():,.2f}"

widgets.interact(simple_backtest, ticker='NVDA', cash=(100, 100000, 50))
```
