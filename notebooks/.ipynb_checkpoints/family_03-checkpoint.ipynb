This code represents a **Monte Carlo Simulation** or "Random Walk" analysis. Instead of testing one specific strategy (like a Moving Average), it tests **1,000 completely random strategies** to see what the "average" outcome looks like for BTC and ETH.

Here is a breakdown of what each part of the script is doing:

### 1. Generating the Randomness

Python

```
n = np.random.randint(10, 101, size=1000).tolist()
```

The code creates a list of 1,000 numbers. Each number (between 10 and 100) represents **how many trades** a strategy will make.

- Strategy 1 might trade 12 times.
    
- Strategy 2 might trade 88 times.
    
- This helps you see if trading _more often_ leads to better or worse results.
    

### 2. Executing the "Monkey" Trades

Python

```
pf = vbt.Portfolio.from_random_signals(price, n=n, init_cash=100, seed=42)
```

VectorBT acts like 1,000 monkeys throwing darts at a board. It generates random entries and exits for both BTC and ETH based on the trade counts defined above. The `seed=42` ensures that if you run the code again, you get the same "random" results.

### 3. Calculating "Expectancy"

Python

```
mean_expectancy = pf.trades.expectancy().groupby(['randnx_n', 'symbol']).mean()
```

**Expectancy** is one of the most important metrics in trading. It answers the question: _"On average, how many dollars do I win (or lose) for every dollar I risk?"_

- **Positive Expectancy:** The strategy makes money over time.
    
- **Negative Expectancy:** The strategy loses money over time (common with random trading due to fees/volatility).
    

### 4. Visualizing the Results

The final scatter plot creates a "Cloud of Outcomes."

- **X-axis (`randnx_n`):** The number of trades made.
    
- **Y-axis (`mean_expectancy`):** The average profit/loss per trade.
    
- **The Goal:** Traders use this to see if any random strategies accidentally performed well, or to prove that random trading on these assets tends to converge toward a specific value (usually near zero or slightly positive for assets in a long-term bull market).
    

---

### Why is this useful?

This is a **Benchmark of Luck.** If you develop a "fancy" AI strategy that has a mean expectancy of 0.05, but a random strategy also has an expectancy of 0.05, it means your AI isn't actually "smart"â€”it's just benefiting from the general upward trend of the market.

````python

import vectorbt as vbt

import numpy as np

symbols = ["BTC-USD", "ETH-USD"]
price = vbt.YFData.download(symbols, missing_index='drop').get('Close')

n = np.random.randint(10, 101, size=1000).tolist()
pf = vbt.Portfolio.from_random_signals(price, n=n, init_cash=100, seed=42)

mean_expectancy = pf.trades.expectancy().groupby(['randnx_n', 'symbol']).mean()
fig = mean_expectancy.unstack().vbt.scatterplot(xaxis_title='randnx_n', yaxis_title='mean_expectancy')
fig.show()
````

````python
import vectorbt as vbt
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- UI Setup ---
ticker_select = widgets.Dropdown(options=['BTC-USD', 'ETH-USD', 'NVDA'], value='BTC-USD', description='Asset:')
fee_slider = widgets.FloatSlider(value=0.001, min=0, max=0.01, step=0.001, readout_format='.3f', description='Fee (0.1%):')
range_slider = widgets.IntRangeSlider(value=[10, 50], min=2, max=100, description='MA Range:')
run_btn = widgets.Button(description="Optimize Strategy", button_style='warning')
out = widgets.Output()

def run_optimization(b):
    with out:
        clear_output()
        print(f"Running optimization for {ticker_select.value}...")
        
        # 1. Download data
        price = vbt.YFData.download(ticker_select.value, period='2y').get('Close')
        
        # 2. Define a range of windows to test (e.g., 10 to 50)
        windows = np.arange(range_slider.value[0], range_slider.value[1])
        
        # 3. Fast way to run all combinations (vectorbt magic)
        fast_ma, slow_ma = vbt.MA.run_combs(price, window=windows, r=2, short_names=['fast', 'slow'])
        
        entries = fast_ma.ma_crossed_above(slow_ma)
        exits = fast_ma.ma_crossed_below(slow_ma)
        
        # 4. Run portfolio with fees
        pf = vbt.Portfolio.from_signals(price, entries, exits, fees=fee_slider.value, init_cash=100)
        
        # 5. Visualize the Heatmap of Total Returns
        returns = pf.total_return()
        fig = returns.vbt.heatmap(
            title=f"Optimization Heatmap: {ticker_select.value}",
            xaxis_title="Slow MA Window",
            yaxis_title="Fast MA Window"
        )
        fig.show()
        
        print(f"Best Combination Total Return: {returns.max()*100:.2f}%")

run_btn.on_click(run_optimization)
display(widgets.VBox([ticker_select, fee_slider, range_slider, run_btn]), out)
````

### Key Concepts in this code:

- **`vbt.MA.run_combs`**: This is the "optimization engine." Instead of a loop, it generates every possible pair of Fast/Slow MAs at once.
    
- **The Heatmap**: In the chart, the "brightest" spots represent the most profitable window combinations.
    
- **Fees**: By adjusting the `fee_slider`, you'll notice the profitable "spots" on the heatmap start to disappear. This is a crucial lesson in quant trading: **over-trading with high fees kills strategies.**
    

### Comparison Table: Strategy vs. Benchmark

When you run the widget above, you'll see how "overfitting" works. A window that worked for NVDA might fail for BTC.

|**Metric**|**Strategy (Optimum)**|**Buy & Hold (Benchmark)**|
|---|---|---|
|**Total Profit**|Found via Heatmap|Usually higher in Bull Markets|
|**Drawdown**|Usually lower (Exits in downtrend)|100% exposure to crashes|
|**Effort**|High (Requires tuning)|Zero|

````python
import vectorbt as vbt
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. UI Elements
symbols_select = widgets.SelectMultiple(
    options=['BTC-USD', 'ETH-USD', 'NVDA', 'AAPL', 'TSLA'],
    value=['BTC-USD', 'ETH-USD', 'NVDA'],
    description='Assets:'
)
fast_ui = widgets.IntSlider(value=10, min=2, max=50, description='Fast MA:')
slow_ui = widgets.IntSlider(value=30, min=10, max=100, description='Slow MA:')
run_btn = widgets.Button(description="Run Analysis", button_style='success')
output_box = widgets.Output()

def run_fixed_analysis(b=None):
    with output_box:
        clear_output()
        symbols = list(symbols_select.value)
        if not symbols: return print("Select an asset!")

        # Set frequency to 'D' (Daily) to fix Sharpe/Calmar warnings
        price = vbt.YFData.download(symbols, period='2y').get('Close')
        
        fast_ma = vbt.MA.run(price, fast_ui.value)
        slow_ma = vbt.MA.run(price, slow_ui.value)
        entries = fast_ma.ma_crossed_above(slow_ma)
        exits = fast_ma.ma_crossed_below(slow_ma)

        # Build Portfolio - Added 'freq' here
        pf = vbt.Portfolio.from_signals(price, entries, exits, init_cash=100, freq='D')
        
        # --- FIXING THE STYLE ERROR ---
        # We use .to_frame().T to ensure we ALWAYS have a DataFrame even for 1 asset
        stats_df = pf.stats()
        if isinstance(stats_df, pd.Series):
            stats_df = stats_df.to_frame().T
        else:
            stats_df = stats_df.T # Transpose so assets are rows

        cols = ['Total Return [%]', 'Benchmark Return [%]', 'Sharpe Ratio', 'Max Drawdown [%]']
        print("--- Strategy Performance Summary ---")
        display(stats_df[cols].style.format("{:.2f}").background_gradient(cmap='RdYlGn'))

        # --- FIXING THE PLOT ERROR ---
        # Explicitly plot the value of each column
        fig = pf.value().vbt.plot()
        fig.update_layout(title="Growth of $100 per Asset", yaxis_title="USD ($)")
        fig.show()

run_btn.on_click(run_fixed_analysis)
display(widgets.VBox([symbols_select, fast_ui, slow_ui, run_btn]), output_box)
````
