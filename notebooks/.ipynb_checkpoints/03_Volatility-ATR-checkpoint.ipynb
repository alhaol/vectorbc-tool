## Concept 3: Average True Range (ATR)

### 1. The Math: "Measuring the Wiggle"

Most people think of volatility as how much the price changes from yesterday's close. But that misses "Gaps" (when a stock opens much higher or lower than it closed).

The **True Range (TR)** is the largest of these three distances:

1. **Current High** minus **Current Low** (Today's "Wiggle").
    
2. **Current High** minus **Yesterday's Close** (The "Gap Up").
    
3. **Current Low** minus **Yesterday's Close** (The "Gap Down").
    

The **ATR** is simply the **Moving Average** of those True Ranges (usually over 14 days).


$$ATR = \text{Average of (True Ranges over 14 days)}$$

---

### 2. The Visualization: The "Safety Buffer"

Think of ATR as a **bubble** around the price.

- **Low ATR:** The stock is moving in tiny steps (quiet market). You can keep your Stop Loss very close.
    
- **High ATR:** The stock is jumping around wildly (nervous market). You need a wide "safety buffer" so you don't get kicked out of a trade by accident.
    

---

### 3. The "Numbers" in Action

If AAPL is at **$200** and its ATR is **$5**, it means on an average day, it moves $5.

- A "safe" Stop Loss might be **2x ATR** ($10).
    
- So, you would sell if the price hits **$190**.
    

---

### 4. VectorBT Implementation

VectorBT has ATR built-in. Let's visualize the price and the ATR below it to see how they react to "chaos."

```python
import vectorbt as vbt
import plotly.graph_objects as go

# 1. Download Data
data = vbt.YFData.download('AAPL', period='6mo')
close = data.get('Close')
high = data.get('High')
low = data.get('Low')

# 2. Calculate ATR using VectorBT
atr = vbt.ATR.run(high, low, close, window=14).atr

# 3. Visualize
fig = vbt.make_subplots(rows=2, cols=1, shared_xaxes=True, 
                        subplot_titles=("Price", "ATR (Market Volatility)"))

fig.add_trace(go.Scatter(x=close.index, y=close, name='Price'), row=1, col=1)
fig.add_trace(go.Scatter(x=atr.index, y=atr, name='ATR', line=dict(color='red')), row=2, col=1)

fig.show()
```

### The Key takeaway for ATR:

The ATR doesn't tell you if the price is going up or down. It only tells you **how much the price is vibrating**.

**Pro Tip:** Smart traders never use a "fixed 5% stop loss." They use an **ATR-based stop**. In a wild market, 5% is too small; in a calm market, 5% is too large.
