Paper trading is "simulated trading." It allows you to execute your **Risk Parity** strategy in real-time markets using play money. Technically, it is the bridge between a historical backtest and risking real capital.

In professional quant finance, paper trading is the **"Forward Test"**—you are proving that your 1.44 Sharpe Ratio wasn't just a fluke of the past.

---

### How it works technically

Technically, paper trading follows a loop: **Data → Logic → Virtual Ledger.**

1. **The Trigger:** A script runs on a schedule (e.g., every morning at 9:30 AM).
    
2. **Live Data Fetch:** The script pulls the latest prices for your basket (AAPL, MSFT, etc.) and your leader (TSLA).
    
3. **Signal Calculation:** It calculates the current RSI, 200-day MA distance, and 20-day volatility.
    
4. **Order Generation:** It calculates the "Target Weights."
    
5. **The Virtual Execution:** Instead of sending an order to a bank, it updates a database or a CSV file acting as your "Virtual Broker."

### Step 3: Handling "Slippage"

The biggest technical difference between paper trading and backtesting is **Slippage**.

- In a backtest, you assume you bought at exactly $150.00.
    
- In live/paper trading, the price might move to $150.05 while your code is running.
    
    A good technical paper trading setup records the Timestamp and the Bid/Ask Spread to ensure the simulation is realistic.

### 3. The "Execution Loop" (Technically)

In a backtest, time moves instantly. In paper trading, you need a **Scheduler**.

Technically, you don't keep the script running 24/7. Instead, you use a **Cron Job** (on Linux/Mac) or **Task Scheduler** (on Windows) to trigger the script once a day, exactly 15 minutes after the market opens.

To make your paper trading accurate, you need your script to calculate volatility using the same "Live" data the broker sees. Alpaca provides a `get_bars` method that is perfect for this.

The code below fetches the history, calculates the **Inverse Volatility Weights**, and prepares the trade.

### The "Live Weight" Calculator

This script replaces the static backtest logic with a live data feed.

Python

```
import alpaca_trade_api as tradeapi
import pandas as pd
import numpy as np

# 1. Setup Keys
API_KEY = "YOUR_PAPER_API_KEY"
SECRET_KEY = "YOUR_PAPER_SECRET_KEY"
BASE_URL = "https://paper-api.alpaca.markets"

api = tradeapi.REST(API_KEY, SECRET_KEY, BASE_URL)

# 2. Settings
basket = ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'META']
window = 20  # Matches our 20-day volatility backtest

def get_live_weights():
    # Fetch daily bars for the last 30 days to ensure we have 20 trading days
    data_list = []
    for symbol in basket:
        bars = api.get_bars(symbol, tradeapi.TimeFrame.Day, "2025-11-20").df
        bars = bars[['close']].rename(columns={'close': symbol})
        data_list.append(bars)
    
    # Combine into one DataFrame
    df = pd.concat(data_list, axis=1).ffill()
    
    # Calculate daily returns
    returns = df.pct_change().dropna()
    
    # Calculate Volatility (Standard Deviation)
    vols = returns.std()
    
    # Calculate Inverse Volatility
    inv_vols = 1 / vols
    
    # Normalize to sum to 1.0 (100% of portfolio)
    weights = inv_vols / inv_vols.sum()
    
    return weights, df.iloc[-1] # Returns weights and latest prices

weights, current_prices = get_live_weights()
print("Target Weights for Today:")
print(weights)
```

### Step 2: The Rebalancing Execution

Now that you have the weights, you need to compare them to what you **currently own** and place the orders to bridge the gap.

Python

```
def rebalance_portfolio():
    weights, prices = get_live_weights()
    account = api.get_account()
    total_cash = float(account.portfolio_value)
    
    for symbol in basket:
        # Calculate how many dollars should be in this stock
        target_value = total_cash * weights[symbol]
        # Calculate target number of shares
        target_qty = target_value // prices[symbol]
        
        # Check current position
        try:
            current_pos = int(api.get_position(symbol).qty)
        except:
            current_pos = 0
            
        diff = target_qty - current_pos
        
        if diff > 0:
            api.submit_order(symbol=symbol, qty=diff, side='buy', type='market')
        elif diff < 0:
            api.submit_order(symbol=symbol, qty=abs(diff), side='sell', type='market')

```

### Key Technical Safeguards

When you transition to this live loop, add these three "Safety Switches":

1. **The Time Gate:** Use `api.get_clock().is_open` to ensure you aren't sending orders on a Sunday.
    
2. **The Rounding Guard:** Stocks must be bought in whole shares (unless your broker supports fractional shares). Use `int()` or `//` to avoid sending an order for `4.5` shares, which will error out.
    
3. **The Error Catch:** Wrap your orders in `try...except` blocks. If Google's API is down, you don't want the whole script to crash before it manages your Amazon position.
