
## Concept 5: Bollinger Bands (The "Rubber Band")

### 1. The Math: Standard Deviation

Bollinger Bands combine the **Moving Average** with **Volatility**.

1. **Middle Band:** A 20-day Simple Moving Average (SMA).
    
2. **Upper Band:** SMA + (2 x Standard Deviation).
    
3. **Lower Band:** SMA - (2 x Standard Deviation).
    

Standard Deviation is just a math term for "how far the price usually wanders from the average."

### 2. The Visualization: The "Squeeze" and "Bulge"

- **The Squeeze:** When the bands get very narrow, it means volatility is low. A "breakout" (big move) is usually coming soon.
    
- **The Bulge:** When the bands are wide apart, the market is chaotic.
    
- **The Reversion:** Statistically, the price stays inside these bands 95% of the time. If it touches the bottom band, itâ€™s like a stretched rubber band ready to snap back to the middle.
    

---

### 4. Why these matter for your Strategy:

If you combine these two new concepts:

- **Buy if:** Relative Strength is rising (it's a leader) AND Price touches the **Lower Bollinger Band** (it's temporarily "stretched" too far down).

### Visualizing Concept 5: Bollinger Bands

Bollinger Bands act like a **flexible envelope** around the price. By measuring how far the price moves away from its average, they tell you if the market is "quiet" or "loud."

---

#### 1. The Anatomy of the Bands

- **The Middle Band (Orange):** This is the baseline trend (20-day Moving Average).
    
- **The Outer Bands (Gray):** These represent the "95% boundaries." Mathematically, price spends about 95% of its time inside this cloud.
    
- **The Cloud (Shaded Area):** This represents the "normal" trading range.
    

---

#### 2. Key Visual Signals

- **The Squeeze:** When the bands pinch together, it means the stock is holding its breath. Volatility is very low, and usually, a "price explosion" (breakout) follows.
    
- **The Bulge:** When the bands expand like a balloon, it means the market has gone wild. High volatility usually leads to exhaustion.
    
- **The Mean Reversion:** Notice how every time the price touches or pokes outside the lower band, it eventually gets "pulled back" toward the orange middle line.
    

---

#### 3. VectorBT Code for Bollinger Bands

In your environment, VectorBT handles the Standard Deviation math automatically. You can use the built-in `vbt.BBANDS` class to see these on your AAPL data:

```python
import vectorbt as vbt
import plotly.graph_objects as go

# 1. Get Data
price = vbt.YFData.download('AAPL', period='6mo').get('Close')

# 2. Calculate Bollinger Bands (Standard: 20-day, 2 Standard Deviations)
bbands = vbt.BBANDS.run(price, window=20, alpha=2)

# 3. Visualize using VectorBT's Plotly integration
# We extract the bands from the bbands object
fig = go.Figure()

# Add the Upper, Middle, and Lower traces
fig.add_trace(go.Scatter(x=price.index, y=bbands.upper, name='Upper Band', line=dict(color='gray', dash='dash')))
fig.add_trace(go.Scatter(x=price.index, y=bbands.middle, name='Middle (SMA)', line=dict(color='orange')))
fig.add_trace(go.Scatter(x=price.index, y=bbands.lower, name='Lower Band', line=dict(color='gray', dash='dash'), fill='tonexty')) # This creates the 'Cloud'

# Add the Price on top
fig.add_trace(go.Scatter(x=price.index, y=price, name='AAPL Price', line=dict(color='blue', width=2)))

fig.update_layout(title="Bollinger Bands: The Price Rubber Band", yaxis_title="Price ($)")
fig.show()
```

### Strategic Insight

The strategy mentioned in Concept 5 is a "Double Confirmation" strategy:

- **Condition 1 (Relative Strength):** Only look at AAPL if it is outperforming the S&P 500 (meaning the "Relative Strength" ratio we just learned is going up).
    
- **Condition 2 (Bollinger Mean Reversion):** Wait for AAPL to touch its **Lower Bollinger Band**.
    

**Why combine them?** You are essentially looking for the "Best Student in Class" (Relative Strength) who just tripped and fell (Bollinger Lower Band). You are betting they will get back up quickly.
