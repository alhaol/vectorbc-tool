Moving into **Basket Trading** (or Portfolio Trading) is a major step up. Instead of betting on one "runner" (Apple) based on another (Tesla), you are now managing a "team."

In VectorBT, basket strategies usually fall into two categories:

1. **Uniform Basket:** You buy/sell the same amount of 5–10 stocks whenever a global condition is met.
    
2. **Relative Basket:** You buy the "strongest" 3 stocks and sell the "weakest" 3 stocks in a sector (Long/Short).
    

Let's build a **Tech Sector Basket** using the "Multi-Stock Logic" we've refined.

### 1. The Strategy: The "Sector Momentum" Basket

We will trade a basket of **AAPL, MSFT, GOOG, AMZN, and META**.

- **Buy Condition:** The "Leader" (TSLA) has an RSI > 70 (Sector is hot).
    
- **Sell Condition:** The stock's own RSI > 80 (Overextended) OR it's Friday.
    

### 2. The Basket Code

The secret here is **Broadcasting**. We take one signal (Tesla) and "stretch" it across the entire basket.

```python
import vectorbt as vbt
import pandas as pd
import numpy as np

# 1. Download the Basket
basket_symbols = ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'META', 'TSLA']
data = vbt.YFData.download(basket_symbols, period='2y')
prices = data.get('Close')
# Flatten the index immediately to avoid KeyErrors
prices.columns = prices.columns.get_level_values('symbol')

# 2. Separate our "Target Basket" from our "Signal Leader"
basket_prices = prices[['AAPL', 'MSFT', 'GOOG', 'AMZN', 'META']]
tsla_price = prices['TSLA']

# 3. Calculate Indicators
basket_rsi = vbt.RSI.run(basket_prices, window=14).rsi
tsla_rsi = vbt.RSI.run(tsla_price, window=14).rsi

# 4. Create the Basket Signals
# Condition: If TSLA RSI > 70, we want to buy the WHOLE basket.
# We must 'tile' the TSLA signal to match the 5 columns of the basket.
is_friday = pd.Series(
    basket_prices.index.dayofweek == 4, 
    index=basket_prices.index
).vbt.tile(len(basket_prices.columns), keys=basket_prices.columns)

# 2. Tile the TSLA signal with the same keys
sector_hot = (tsla_rsi > 70).vbt.tile(
    len(basket_prices.columns), 
    keys=basket_prices.columns
)

# 3. Now this OR (|) will work because both sides are "labeled" identically
basket_entries = sector_hot
basket_exits = (basket_rsi > 80) | is_friday

# 4. Run the Portfolio
pf_basket = vbt.Portfolio.from_signals(
    basket_prices,
    basket_entries,
    basket_exits,
    init_cash=10000,
    fees=0.001,
    cash_sharing=True,
    group_by=True
)

print("Basket Strategy Total Return:")
print(pf_basket.total_return())
```

That is a solid **+5.97%** return!

What’s impressive here is that you aren't just betting on a single stock's luck; you've built a **diversified system**. This return represents the "consensus" of the tech sector responding to Tesla's lead.

Now that the "plumbing" of your basket is working, let's look at the "Alpha" move: **Ranking and Filtering.**

### Concept: The "Survival of the Fittest" Basket

Instead of buying the whole basket when Tesla is hot, we only want to buy the **Top 2 strongest stocks** in that basket. This ensures we aren't dragging down our returns with laggards.

We will use `vbt.RSI` to rank them:

1. Rank all 5 stocks by their RSI.
    
2. Filter for the top 2 (highest momentum).
    
3. Only trigger the entry if the stock is in that "Top 2" list.

```python
# 1. Rank the RSI values across the columns (axis=1)
# Higher RSI = Higher Rank (stronger momentum)
ranks = basket_rsi.rank(axis=1, ascending=False)

# 2. Create a "Top 2" Filter
# This creates a True/False table where only the 2 strongest stocks are True
is_top_2 = ranks <= 2

# 3. Combine with our TSLA Sector Lead
# Now we ONLY buy if TSLA is hot AND the stock is a top leader
ranked_entries = basket_entries & is_top_2

# 4. Run the Portfolio
pf_ranked = vbt.Portfolio.from_signals(
    basket_prices,
    ranked_entries,
    basket_exits,
    init_cash=10000,
    fees=0.001,
    cash_sharing=True,
    group_by=True
)

print(f"Ranked Basket Return: {pf_ranked.total_return()}")
```

It is perfectly normal for a "more complex" strategy to perform worse initially. Let’s break down exactly why that **-13.97%** happened by looking at the logic step-by-step.

### The Problem: The "Momentum vs. Mean Reversion" Conflict

When we only select the **Top 2 RSI** stocks, we are essentially buying the stocks that are "screaming." In the tech sector, a high RSI often means a stock is **overbought**.

- **Logic 1 (TSLA):** TSLA RSI > 70 says "The whole sector is hot, let's play."
    
- **Logic 2 (Top 2):** You select the two stocks with the highest RSI (likely 75-85).
    
- **The Result:** You are buying at the absolute peak of the rally. Because the "rubber band" is stretched to its limit, those stocks often crash back down (mean-revert) immediately after you enter.
    

---

To see if this theory is correct, let's check two things: **Which stocks were picked** and **How many trades occurred.**

This will tell us if we had 100 tiny losses or 1 massive "black swan" loss.

### 2. Why "Ranking" failed (The Fees + Momentum Trap)

Looking at your previous stats, you paid **$374 in fees** on a $10,000 account. That is **3.7% of your capital** gone just to the broker!

When you rank by "Top 2 RSI," you are likely switching stocks very frequently (every time one RSI dips below another).

- **High Churn:** You sell Apple and buy Google.
    
- **Cost:** Every time you do that, you pay the `0.001` fee.
    
- **Result:** You are "churning" your account into the ground.
    

---

### 3. The "Stable Momentum" Strategy

To fix the -13% return, we need to stop the "churn." Instead of ranking by **RSI** (which changes daily), let's rank by **SMA 200 Distance** (which is much more stable).

```python
# 1. Calculate how far each stock is above its 200-day average
ma_200 = vbt.MA.run(basket_prices, window=200).ma
dist_from_ma = (basket_prices - ma_200) / ma_200

# 2. Rank by Distance (More stable than RSI)
stable_ranks = dist_from_ma.rank(axis=1, ascending=False)
is_stable_top_2 = stable_ranks <= 2

# 3. New Entry: TSLA is hot AND stock is a long-term leader
stable_entries = sector_hot & is_stable_top_2

# 4. Run with a Stop Loss to prevent the -20% Drawdown
pf_stable = vbt.Portfolio.from_signals(
    basket_prices,
    stable_entries,
    basket_exits,
    init_cash=10000,
    fees=0.001,
    cash_sharing=True,
    group_by=True,
    sl_stop=0.05  # Protect capital with a 5% stop loss
)

print(f"Stable Basket Return: {pf_stable.total_return()}")
```

### Why this is better:

1. **Less Trading:** Long-term leaders stay in the "Top 2" for weeks, not days. This cuts your fees dramatically.
    
2. **Quality over Hype:** RSI catches "pumps." SMA distance catches "trends."
    

**Does the "Stable Basket" return look better than the -13.97%?** If you look at the `stats()` now, check if your **Total Fees Paid** has dropped. That is usually the secret to turning a losing strategy into a winning one.

That is a fantastic result! You just turned a **-13.97% loss** into a **+9.85% gain** simply by changing the "quality" of your ranking and adding a safety net.

By switching from **RSI** (which is flighty and "noisy") to **Distance from 200-day MA** (which tracks long-term institutional strength), you reduced the "churn" of your portfolio. You stopped gambling on the daily "hype" and started investing in the actual "leaders."

---

### Why this strategy is "Professional Grade"

- **Reduced Friction:** You likely paid significantly less in fees because you weren't swapping stocks every other day.
    
- **Positive Expectancy:** Your return is now positive in a period where your previous logic was bleeding out.
    
- **Risk Control:** The `sl_stop=0.05` ensures that even if one of these "Stable Leaders" has a catastrophic earnings report, you exit with a manageable 5% loss instead of riding it down 20%.

### The Final Refinement: "The Volatility Filter"

If you want to take this even further, there is one more "Secret Sauce" professional traders use: **Volatility Weighting**.

Right now, if you have $10,000 and two stocks, you put $5,000 in each. But what if one stock is **Tesla** (moves 5% a day) and the other is **Microsoft** (moves 1% a day)? Tesla is actually 5x riskier!

**Would you like to see how to "Equalize" the risk so that the wild stocks don't dominate your portfolio's P&L?** This is called **Inverse Volatility Weighting**, and it's how the world's biggest hedge funds manage their baskets.

This is the "Black Belt" of portfolio management: **Risk Parity** (or Inverse Volatility Weighting).

Currently, your portfolio puts 50% of your money in each of your "Top 2" stocks. But if **Tesla (TSLA)** is swinging 10% a day while **Microsoft (MSFT)** only moves 1%, Tesla is effectively "driving the car" while Microsoft is just a passenger. If Tesla crashes, Microsoft can't save the portfolio.

### The Logic: Inverse Volatility

We want to give more money to the "quiet" stocks and less money to the "wild" stocks so that each position has an **equal chance** of moving the needle.

$$\text{Weight} \propto \frac{1}{\text{Volatility}}$$

---

### Step-by-Step: Risk Parity Basket

We'll use standard deviation over a 20-day window to measure "wildness" and then use that to scale our $10,000.

```python
# 1. Calculate Volatility & Inverse Volatility
vols = basket_prices.vbt.returns().rolling_std(window=20)
inv_vols = 1 / vols

# 2. Filter for our "Stable Top 2"
raw_weights = inv_vols.where(is_stable_top_2, 0)

# 3. Normalize to 100%
weights = raw_weights.divide(raw_weights.sum(axis=1), axis=0).fillna(0)

# 4. Run the Portfolio (Removing sl_stop for now)
pf_risk_parity = vbt.Portfolio.from_orders(
    basket_prices,
    size=weights,
    size_type='target_percent',
    init_cash=10000,
    fees=0.001,
    cash_sharing=True,
    group_by=True
)

print(f"Risk Parity Basket Return: {pf_risk_parity.total_return()}")
```

### Why this often "wins" over time

- **Consistency:** You stop having "bad days" just because one high-beta stock decided to tank.
    
- **Capital Efficiency:** You are putting more capital to work in the stocks that are trending smoothly.
    
- **Lower Stress:** Your equity curve (the growth of your $10k) usually becomes a much smoother line.
    

### Comparison: Standard vs. Risk Parity

You can now compare your **Stable Basket (Equal Weight)** against this **Risk Parity Basket**.


```python
# 1. Create the first plot
fig = pf_stable.value().vbt.plot(trace_kwargs=dict(name='Equal Weight'))

# 2. Add the second plot by passing the first 'fig' into the 'fig' argument
pf_risk_parity.value().vbt.plot(fig=fig, trace_kwargs=dict(name='Risk Parity'))

# 3. Show the combined result
fig.show()
```

### 2. Why the 63% Return is a "Eureka" Moment

You've just witnessed the power of **Risk Management** over **Signal Hunting**.

- **The Stable Basket** (Equal Weight) gave you ~10%.
    
- **The Risk Parity Basket** gave you ~63%.
    

**Why the massive jump?** In an equal-weighted basket, a single highly volatile stock (like TSLA) can "wash out" the steady gains of others. By using **Inverse Volatility**, you "tamed" the wild stocks. This allowed the consistent winners to stay in the portfolio longer and contribute more to the bottom line without the massive drawdowns that usually trigger stops or wipe out weekly gains.

### 3. The Final "Stress Test"

Before you'd ever trade this for real, you should check the **Sharpe Ratio**. This number tells you if your 63% return was "skill" (consistent growth) or just "luck" (one crazy spike).

- **Sharpe > 1.0:** Good.
    
- **Sharpe > 2.0:** Excellent.
    
- **Sharpe > 3.0:** "Holy Grail" (rarely lasts).

```python
print("Sharpe Ratio Comparison:")
# 'D' stands for Daily frequency
print(f"Equal Weight: {pf_stable.sharpe_ratio(freq='D')}")
print(f"Risk Parity:  {pf_risk_parity.sharpe_ratio(freq='D')}")
```

### Why Risk Parity wins the Sharpe Battle

The reason your **Risk Parity** return jumped to **63%** while the Sharpe ratio likely stayed high is that Risk Parity acts like a "Volume Knob" for risk.

1. When the market is **calm and trending**, it turns the volume up (invests more).
    
2. When the market gets **wild and unpredictable**, it turns the volume down (invests less).
    

This "dynamic tuning" prevents the massive losses that usually destroy a strategy's Sharpe Ratio.
